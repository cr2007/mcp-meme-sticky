```python
import sys
sys.path.append('..')
```

```python
from mediapipe.tasks import python
from mediapipe.tasks.python import text
import pickle
```

```python
from resources.db_metadata import templates_metadata
```

```python
PATH = '../resources/embedder.tflite'
```

```python
base_options = python.BaseOptions(model_asset_path=PATH)
l2_normalize, quantize = True, False
options = text.TextEmbedderOptions(
    base_options=base_options, l2_normalize=l2_normalize, quantize=quantize)
embedder = text.TextEmbedder.create_from_options(options)
```

```python
embeddings = {}

for k,v in templates_metadata.items():
    embeddings[k] = embedder.embed(v).embeddings[0]
```

```python
embeddings
```

```python
with open('../resources/db_embeddings.pkl', 'wb') as f:
    pickle.dump(embeddings, f)
```
